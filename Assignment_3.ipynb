```python
lst = [1,2,2,3,2,3,4,5]
count = {}
for i in lst:
  if i in count:
    count[i]+=1
  else:
    count[i]=1
temp = 0
for k,v in count.items():
  if v > temp:
    temp = v
    result = k
print("The given list of numbers is: ",lst)
print("The number of each element is: ",count)
print("The number of element with highest repetation is: ",result,"and the number of time it repeated is:",count[result])
```
